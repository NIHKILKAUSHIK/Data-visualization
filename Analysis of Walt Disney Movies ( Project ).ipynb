#Visualization of trends using variable release_date

```python
df['release_year'] = pd.to_datetime(df['release_date']).dt.year
```

This line of code is creating a new column called 'release_year' in the DataFrame 'df'.
It is using the 'pd.to_datetime()' function to convert the values in the 'release_date' column to datetime format.
Then, it is using the 'dt.year' attribute to extract the year from each datetime value and assign it to the 'release_year'
column.

```python
genre_yearly_gross = df.groupby(['release_year', 'genre'])['total_gross'].sum().reset_index()
```

The code is grouping the DataFrame `df` by the columns `release_year` and `genre`. It then calculates the sum of the `total_gross` column for each group. The result is a new DataFrame `genre_yearly_gross` with three columns: `release_year`, `genre`, and `total_gross`.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=genre_yearly_gross, x='release_year', y='total_gross', hue='genre')
plt.title('Genre Gross Earnings Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Gross Earnings')
plt.legend(title='Genre', loc='upper left', bbox_to_anchor=(1, 1))
```

The code is creating a line plot using the seaborn library.

#Decade variable for further trend analysis

```python
df['release_year'] = pd.to_datetime(df['release_date']).dt.year
df['decade'] = (df['release_year'] // 10) * 10
```

The code is adding two new columns to the DataFrame `df`.

```python
decade_genre_counts = df.groupby(['decade', 'genre']).size().reset_index(name='count')
```

The code is grouping the DataFrame `df` by the columns 'decade' and 'genre'. It then calculates the size of each group and resets the index of the resulting DataFrame. The resulting DataFrame is assigned to the variable `decade_genre_counts`, with the column name 'count'. This code is essentially counting the number of occurrences of each combination of 'decade' and 'genre' in the original DataFrame.

```python
plt.figure(figsize=(10, 6))
plt.bar(decade_genre_counts['decade'], decade_genre_counts['count'])
plt.xlabel('Decade')
plt.ylabel('Count')
plt.title('Count of Movies by Genre in Each Decade')
plt.xticks(rotation=45)
plt.show()
```

This code is creating a bar chart using the `matplotlib.pyplot` library.

#Using movie_title lets classify movies based on their franchise.

```python
franchise_dict = {
    "Toy Story": "Pixar",
    "Pirates of the Caribbean": "Disney",
    "Cars": "Pixar",
    "Prince of Persia": "Other",
    "Muppets": "Disney",
    "Marvel": "Marvel",
    "Star Wars": "Star Wars"
}
```

The code is creating a dictionary called `franchise_dict` that maps movie franchises to their respective production companies. Each key-value pair in the dictionary represents a movie franchise and its corresponding production company.

```python
df['franchise'] = df['movie_title'].map(franchise_dict)
```

This line of code is creating a new column called 'franchise' in the DataFrame 'df'. The values in this new column are derived from the 'movie_title' column by mapping each movie title to its corresponding value in the 'franchise_dict' dictionary.

```python
franchise_counts = df['franchise'].value_counts()

plt.figure(figsize=(10, 6))
franchise_counts.plot(kind='hist')
plt.title('Movie Distribution by Franchise')
plt.xlabel('Franchise')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

This code is creating a bar chart to visualize the distribution of movies by franchise.

#Let's see if movies which have a franchise performed better to movies who don't have franchise

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='franchise', y='total_gross', data=df)
plt.title('Comparison of Earnings for Movies with and without a Franchise')
plt.xlabel('Franchise')
plt.ylabel('Earnings')
plt.show()
```

The code is creating a boxplot to compare the earnings of movies with and without a franchise.

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='total_gross', hue='franchise', element='step', common_norm=False, stat='density')
plt.title('Distribution of Earnings for Movies with and without a Franchise')
plt.xlabel('Earnings')
plt.ylabel('Density')
plt.show()
```

The code is creating a histogram plot using the seaborn library. It is plotting the distribution of earnings for movies with and without a franchise. The x-axis represents the earnings, and the y-axis represents the density. The `hue` parameter is used to differentiate between movies with and without a franchise. The `element` parameter is set to 'step' to create a step-like histogram. The `common_norm` parameter is set to False to normalize the histograms separately. The `stat` parameter is set to 'density' to display the density instead of the count. The `plt.title`, `plt.xlabel`, and `plt.ylabel` functions are used to set the title and labels for the plot. Finally, `plt.show()` is used to display the plot.

```python
avg_earnings_by_franchise = df.groupby('franchise')['total_gross'].mean()
plt.figure(figsize=(10, 6))
avg_earnings_by_franchise.plot(kind='hist')
plt.title('Average Earnings for Movies with and without a Franchise')
plt.xlabel('Franchise')
plt.ylabel('Average Earnings')
plt.xticks(rotation=0)
plt.show()
```

This code is calculating the average earnings for movies with and without a franchise. It groups the data by the 'franchise' column and calculates the mean of the 'total_gross' column for each group. Then, it creates a bar plot to visualize the average earnings by franchise. The plot has a title, x and y labels, and the x-axis labels are not rotated. Finally, it displays the plot.

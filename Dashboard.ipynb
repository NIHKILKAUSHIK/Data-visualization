```python
from dash import Dash,dcc,html,Input,Output,dash_table
import plotly.express as px
import pandas as pd
```

```python

df=pd.read_csv("C:/Users/91863/Downloads/disney_movies_total_gross.csv")
#df=px.data.gapminder()
```

```python
df.isnull().sum()
```

```python
genre = (df['genre'].isnull().sum()/len(df['genre']))*100
genre
```

```python
mpaa_rating = (df['mpaa_rating'].isnull().sum()/len(df['mpaa_rating']))*100
mpaa_rating
```

```python
df['genre']=df['genre'].fillna(df['genre'].mode()[0])
df.genre
```

```python
df['mpaa_rating']=df['mpaa_rating'].fillna(df['mpaa_rating'].mode()[0])
df.mpaa_rating
```

```python
df.isnull().sum()
```

```python
df['release_year'] = pd.to_datetime(df['release_date']).dt.year
```

```python
df['decade'] = (df['release_year'] // 10) * 10
```

```python
decade_genre_counts = df.groupby(['decade', 'genre']).size().reset_index(name='count')
```

```python
franchise_dict = {
    "Toy Story": "Pixar",
    "Pirates of the Caribbean": "Disney",
    "Cars": "Pixar",
    "Prince of Persia": "Other",
    "Muppets": "Disney",
    "Marvel": "Marvel",
    "Star Wars": "Star Wars"
}
```

```python
df['franchise'] = df['movie_title'].map(franchise_dict)
```

```python
franchise_counts = df['franchise'].value_counts()
```

```python
app=Dash(__name__)
```

```python
app.layout=html.Div([
    dash_table.DataTable(data=df.to_dict('records'),page_size=10),
    html.H4('Analysis of Walt Disney Movies'),
    html.P('Select Graph'),
    dcc.Dropdown(
        id='selection',
        options=['Line Plot - Genre Gross Earnings Over the Years',
                 "Line Plot - Genre Inflation-Adjusted Gross Earnings Over the Years",
                 "Line Plot - Inflation-Adjusted Gross Earnings by MPAA Rating Over the Years",
                 "Bar Chart - Count of Movies by Genre in Each Decade",
                 "Box Plot - Comparison of Earnings for Movies with and without a Franchise"],
        value='Line Plot - Genre Gross Earnings Over the Years',
    ),

dcc.Loading(dcc.Graph(id='graph'),type='cube')
])
@app.callback(
    Output("graph","figure"),
    Input('selection','value'),
)
def dispaly_animated_graph(selection):
    animations= {
        "Line Plot - Genre Gross Earnings Over the Years":px.line(
            df,x='release_year',
            y='total_gross',
            color='genre',
        ),
        "Line Plot - Genre Inflation-Adjusted Gross Earnings Over the Years":px.line(
            df,x='release_year',
            y='inflation_adjusted_gross',
            color='genre',
        ),
        "Line Plot - Inflation-Adjusted Gross Earnings by MPAA Rating Over the Years":px.line(
            df,x='release_year',
            y='inflation_adjusted_gross',
            color='mpaa_rating',
#         ),
#         "Bar Chart - Count of Movies by Genre in Each Decade":px.bar(
#             decade_genre_counts,x=decade_genre_counts['count'],
# #             y=decade_genre_counts['count'],
#             color=decade_genre_counts['decade'],
#         ),
#         "Box Plot - Comparison of Earnings for Movies with and without a Franchise":px.box(
#             df,x='total_gross',
#             y='franchise',
        )
    }
    return animations[selection]
```

```python

if __name__ == '__main__':
    app.run(debug=True)
```
